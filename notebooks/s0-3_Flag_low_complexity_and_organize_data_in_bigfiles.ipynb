```python
# Write a flagging processing function
def flag_loom_file(filename, plot_path="/data2/loom_files/development/autogenerated_figures/complexity_flags/"):
    ds = loompy.connect(filename)
    M = ds[:]
    total_molecules = M.sum(0)
    low_expression_flag = QC.low_expression_flagger(total_molecules, thresh=1600)
    ds.set_attr("Low_expression_flag", low_expression_flag, axis=1)
    M = M[:,low_expression_flag == False]
    total_molecules, expressed_genes = M.sum(0),(M>0).sum(0)
    temp_flag = QC.low_complexity_flagger(total_molecules, expressed_genes)
    low_complexity_flag = np.zeros_like(low_expression_flag, dtype=temp_flag.dtype)
    low_complexity_flag[low_expression_flag==False] = temp_flag
    ds.set_attr("Low_complexity_flag", low_complexity_flag, axis=1)
    print("Flagging of %s worked!" % filename.split("/")[-1])
    
    if plot_path:
        #Save a plot to visualize later
        fig = plt.figure()
        plt.scatter( total_molecules, expressed_genes, s=6, alpha=0.5,lw=0, c=array(["b","r"])[temp_flag.astype(int)] );
        plt.title("Complexity flag for %s" % filename.split("/")[-1])
        plt.ylim(-50,)
        plt.xlabel("log( Total Molecules )")
        plt.ylabel("log( Expressed Genes )")
        plt.tight_layout() # !!!!Might not work depending the backend!!!!
        plt.savefig(os.path.join(plot_path,"%s.png" % filename.split("/")[-1].rstrip(".loom")), bbox_inces="tight")
        fig.clf()
```

```python
len(info_files)
```

```python
# Check if thre row attributes correspond
for info_f in info_files:
    files_to_check = list(info_f[-1])
    f0 = files_to_check[0]
    ds0 = loompy.connect(f0)
    fs = files_to_check[1:]
    for f in fs:
        ds = loompy.connect(f)
        if np.alltrue( ds.row_attrs["Accession"] == ds0.row_attrs["Accession"] ):
            pass
        else:
            raise ValueError("Row attributes do not correspond")
```

```python
import multiprocessing

# Parallel processing
pool = multiprocessing.Pool(processes=7)
try:
    for info_f in info_files:
        pool.apply_async(combine_using_info,(info_f,))
    pool.close()
    pool.join()
except KeyboardInterrupt:
    pool.terminate()
```
